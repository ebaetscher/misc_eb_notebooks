$N_+ / N_- = e^ {\Delta E / kT}$

Boltzmann in 1868 is credited with first studying this function and its application to statistical mechanics and thermodynamics. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gamma_1H = 42.577478518E6 #Hz per Tesla
k_boltz = 1.380649E-23 #Joules per Kelvin
h_planck = 6.626E-34 #J per Hz
t_kelvin = 293.15 # 0C


def boltz_dist_proton_for_tesla(tesla = np.linspace(-10,10,101),
                                gamma = 42.577478518E6,
                                k_boltz = 1.380649E-23,
                                h_planck = 6.626E-34,
                                t_kelvin = (293.15 + 37),
                                make_plot = True):
    
    
    delta_E = (tesla * gamma * h_planck)
    boltz_y_val_1 = np.exp(-1 * delta_E / (k_boltz * t_kelvin))
    
    
    if make_plot == True:
        plt.figure(figsize=(7,7))
        plt.plot(tesla, boltz_y_val_1)
        plt.vlines(0.0, boltz_y_val_1.min(), boltz_y_val_1.max(), ls = '--')
        plt.hlines(1.0, tesla.min(), tesla.max(), ls = '--')
        plt.xlabel('field strength (T)')
        plt.ylabel('ratio of spins in beta (higher energy) state to \n spins in alpha (lower energy) state')
        plt.ticklabel_format(useOffset=False, style='plain')
        plt.title('Boltzmann dist - ' + str(t_kelvin) + ' Kelvin - ' + str(gamma) + ' Hz/Tesla' )
        plt.show()
    
    return(tesla, boltz_y_val_1)

```

```python
tsla_rtn_0, blotz_rtn_0 = boltz_dist_proton_for_tesla()
```

```python
tsla_rtn_0, blotz_rtn_0 = boltz_dist_proton_for_tesla(t_kelvin = 4.0)
```

```python
tsla_rtn_0, blotz_rtn_0 = boltz_dist_proton_for_tesla(tesla = np.linspace(-10,1000000,101))
```

In the above plot, we go out to the 'crazy' field strength of 1-million Tesla to demonstrate that eventually the apparent linear relationship between spin polarization and field strength is no longer a good approximation at extremely high fields. Indeed, given that the strongest MRI systems in the world produce fields on the order of 10 Tesla, field strengths would need to be increased by 4 orders of magnitude for this difference from linearity to become important. Such fields of 100,000 Tesla would surely be intolerable for human or live-creature use, and are so far beyond the state of the art as to be technology currently in the realm of science-fiction. In astronomy, the magnetic fields near the surface of neutron stars can reach and exceed 10^6 Tesla, so fields of such magnitude are physically relevant however.  


Now, the Boltzmann distrubution in this case describes a spin system at thermal equilibrium. It turns out that there are ways to prepare the spin system such that this model of themal equilibrium does not hold true, and spin polarizations far in excess of the meager room-temperature values can be achived. These techniques are generally called "hyperpolarization". These hyperpolarization techniques are currently rather difficult to implement and are rarely used in the clinical setting. It would be an incredible breakthrough to be able to hyperpolarize the nuclear spins in a sample or person in a simpler way than is presently done.
