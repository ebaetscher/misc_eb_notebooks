## SNR as a function of T1 and Flip Angle
for a gradient echo MRI pulse sequence

-E.Baetscher 2021-02-08

    Lipid T1 ~ 0.4s 
    Water T1 range ~ 1.3 to 1.7

    TR = 10ms

    S = ((1 - e ^ -TR / T1))/(1 - (cos alpha) e ^ -TR / T1) * (e ^ -TE / T2_star)

```python
import numpy as np
import matplotlib.pyplot as plt

def signal_from_T1_alpha_TR(t1_domain, alpha_deg, rep_time, echo_time, t2_star):
    
    alpha = (alpha_deg / 360.0) * (np.pi * 2.0)
    #signal = (np.sin(alpha) * (1 - np.exp(-1 * rep_time / t1_domain)))/(1 - (np.cos(alpha) * np.exp(-1 * rep_time / t1_domain)))
    signal = ((1 - np.exp(-1 * rep_time / t1_domain))/(1 - (np.cos(alpha) * np.exp(-1 * rep_time / t1_domain)))) * (np.sin(alpha) * np.exp(-1 * echo_time / t2_star))
    return signal
```

```python
%matplotlib inline

flip_angle_arr = np.linspace(0,40.0, 101)
sig_1p4 = signal_from_T1_alpha_TR(1.4, flip_angle_arr, 0.01, 0.003, 0.100)
sig_1p5 = signal_from_T1_alpha_TR(1.5, flip_angle_arr, 0.01, 0.003, 0.100)
sig_1p6 = signal_from_T1_alpha_TR(1.6, flip_angle_arr, 0.01, 0.003, 0.100)
sig_0p4 = signal_from_T1_alpha_TR(0.4, flip_angle_arr, 0.01, 0.003, 0.100)
#sig_0p08 = signal_from_T1_alpha_TR(0.08, flip_angle_arr, 0.01, 0.003, 0.100)

plt.figure()
plt.plot(flip_angle_arr, sig_1p4, label='1.4')
plt.plot(flip_angle_arr, sig_1p5, label='1.5')
plt.plot(flip_angle_arr, sig_1p6, label='1.6')
plt.plot(flip_angle_arr, sig_0p4, label='0.4')
#plt.plot(flip_angle_arr, sig_0p08)
plt.xlabel('flip-angle (degrees)')
plt.ylabel('signal (a.u.)')
plt.legend()
plt.show()
```

```python
%matplotlib inline
#%matplotlib qt

t1_dom = np.linspace(0.1, 1.8, 101)
sig_0 = signal_from_T1_alpha_TR(t1_dom, 5.0, 0.01, 0.003, 0.100)
sig_1 = signal_from_T1_alpha_TR(t1_dom, 6.0, 0.01, 0.003, 0.100)
sig_2 = signal_from_T1_alpha_TR(t1_dom, 7.0, 0.01, 0.003, 0.100)
sig_3 = signal_from_T1_alpha_TR(t1_dom, 8.0, 0.01, 0.003, 0.100)
sig_4 = signal_from_T1_alpha_TR(t1_dom, 9.0, 0.01, 0.003, 0.100)
sig_5 = signal_from_T1_alpha_TR(t1_dom, 10.0, 0.01, 0.003, 0.100)


plt.figure()
plt.plot(t1_dom, sig_0, label='5.0')
plt.plot(t1_dom, sig_1, label='6.0')
plt.plot(t1_dom, sig_2, label='7.0')
plt.plot(t1_dom, sig_3, label='8.0')
plt.plot(t1_dom, sig_4, label='9.0')
plt.plot(t1_dom, sig_5, label='10.0')
plt.xlabel('T1 (seconds)')
plt.ylabel('signal (a.u.)')
plt.legend()
plt.show()
```
