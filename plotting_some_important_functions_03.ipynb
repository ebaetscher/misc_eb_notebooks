# Some basic mathematical functions, plotted

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
```

```python
#We define a nice plotting function

def plot_w_spines_at_zero(x_arr, y_arr,
                          title=None,
                          x_tick_spacing=1.0,
                          y_tick_spacing=1.0,
                          x_lims=[-10,10],
                          y_lims=[-10,10]):
    
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1)
    
    #ax.set_title('centered spines')
    

    ax.plot(x_arr, y_arr, linewidth=3, color='r')
    ax.spines['left'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('zero')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    
    ax.set_xlim(x_lims)
    ax.set_ylim(y_lims)
    
    x_loc = plticker.MultipleLocator(base=x_tick_spacing) # this locator puts ticks at regular intervals
    y_loc = plticker.MultipleLocator(base=y_tick_spacing)
    ax.xaxis.set_major_locator(x_loc)
    ax.yaxis.set_major_locator(y_loc)
    ax.grid(ls='--')
    
    plt.show()
    
    return
```

```python
#Now we set a domain from -10 to 10

x_0 = np.linspace(-10,10,1001)
```

```python
#first let's do the identity function
y_0 = (x_0 * 1.0)

plot_w_spines_at_zero(x_0, y_0)
```

```python
# 1 over x
y_1 = (1.0 / x_0)

plot_w_spines_at_zero(x_0, y_1)
```

```python
#now a quadratic function
y_2 = x_0**2

plot_w_spines_at_zero(x_0, y_2)
```

```python
#a cubic function
y_3 = x_0**3

plot_w_spines_at_zero(x_0, y_3)
```

```python
#exponential function
y_4 = np.exp(x_0)

plot_w_spines_at_zero(x_0, y_4)
```

```python
#natural log of x
y_5 = np.log(x_0)

plot_w_spines_at_zero(x_0, y_5)
```

```python
#square root of x
y_6 = np.sqrt(x_0)

plot_w_spines_at_zero(x_0, y_6)
```

```python
#Gaussian distribution
mu_0 = 0.0
sigma_0 = 1.0

y_7 = ((1 / (sigma_0 * np.sqrt(np.pi * 2.0))) * np.exp(-0.5 * ((x_0 - mu_0) / sigma_0) ** 2))

plot_w_spines_at_zero(x_0, y_7, y_lims=[-1.0,1.0], y_tick_spacing=0.1)
```

```python
#Lorentzian distribution

y_8 = (1 / ((np.pi * sigma_0) * (1 + (x_0 / sigma_0)**2)))

plot_w_spines_at_zero(x_0, y_8, y_lims=[-1.0,1.0], y_tick_spacing=0.1)
```

```python
#sine wave

y_9 = (np.sin(x_0))

#tan function
y_10 = (np.tan(x_0))

plot_w_spines_at_zero(x_0, y_9)

plot_w_spines_at_zero(x_0, y_10)
```
